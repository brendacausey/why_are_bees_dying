```python
#Merged the two tables
bees = pd.merge(
    bees_lost,
    bees_stressors,
    on=["year", "quarter", "state", "state_code"],
    how="left")

print(bees.head())
```

```python
#Kept getting a Key Error telling me that the 'percent_lost' column didn't exist. 
#Ran this to figure out if there was anything weird about this column. 
#Sure enough, there is a space after it. So, added a space and it ran fine.

print(bees.columns)

if 'percent_lost ' in bees.columns:
    print(bees['percent_lost '])
else:
    print("Column does not exist.")
```

```python
#Are bee colony losses increasing over time?

loss_trend = bees.groupby("year")["percent_lost "].mean().reset_index()

print(loss_trend)
```

```python
#Line Chart

plt.figure()
plt.plot(loss_trend["year"], loss_trend["percent_lost "], marker='o')
plt.title("Average Bee Colony Loss Over Time")
plt.xlabel("Year")
plt.ylabel("Average Percent Lost")
plt.show()
```

```python
#Which states experience the highest losses?

state_losses = (
    bees.groupby("state")["percent_lost "]
    .mean()
    .sort_values(ascending=False)
    .reset_index())

print(state_losses.head(10))
```

```python
#Bar Chart

plt.figure()
plt.bar(state_losses["state"].head(10), state_losses["percent_lost "].head(10))
plt.xticks(rotation=45)
plt.title("Top 10 States by Average Colony Loss")
plt.ylabel("Percent Lost")
plt.show()
```

```python
#What factors correlate most strongly with colony collapse?

correlation_columns = [
    "percent_varroa_mites",
    "percent_other_pests",
    "percent_diseases",
    "percent_pesticides",
    "percent_other",
    "percent_unknown"]

correlations = bees[["percent_lost "] + correlation_columns].corr()

print(correlations["percent_lost "].sort_values(ascending=False))
```

```python
#Pesticides

pesticide_corr = bees["percent_lost "].corr(bees["percent_pesticides"])
print("Correlation between pesticide exposure and colony loss:", pesticide_corr)
```

```python
#Scatter plot

plt.figure()
plt.scatter(bees["percent_pesticides"], bees["percent_lost "])
plt.title("Pesticide Exposure vs Colony Loss")
plt.xlabel("Percent Colonies Affected by Pesticides")
plt.ylabel("Percent Colonies Lost")
plt.show()
```

```python
#Varroa mite

varroa_corr = bees["percent_lost "].corr(bees["percent_varroa_mites"])
print("Correlation with Varroa mites:", varroa_corr)
```

```python
#Scatter plot

plt.figure()
plt.scatter(bees["percent_varroa_mites"], bees["percent_lost "])
plt.title("Varroa Mites vs Colony Loss")
plt.xlabel("Percent Colonies Affected by Varroa Mites")
plt.ylabel("Percent Colonies Lost")
plt.show()
```

```python
commercial_states = ["California","North Dakota","Florida","Texas","South Dakota"]

commercial = bees[bees["state"].isin(commercial_states)]
non_commercial = bees[~bees["state"].isin(commercial_states)]

commercial_loss = commercial["percent_lost "].mean()
non_commercial_loss = non_commercial["percent_lost "].mean()
print(commercial_loss)
print(non_commercial_loss)
```

```python
#Seasonal losses

quarter_losses = bees.groupby("quarter")["percent_lost "].mean()

print(quarter_losses)
```

```python
#Bar Chart

plt.figure()
plt.bar(quarter_losses.index, quarter_losses.values)
plt.title("Average Colony Loss by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Percent Lost")
plt.show()
```

```python
#Can we see seasonal patterns?

seasonal_pattern = bees.pivot_table(
    values="percent_lost ",
    index="quarter",
    columns="year",
    aggfunc="mean")

print(seasonal_pattern)
```
